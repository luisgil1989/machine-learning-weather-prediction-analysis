```python
# List of cities to be removed
cities_to_remove = ['GDANSK', 'ROMA', 'TOURS']

# Filter out columns that contain the names of the cities to remove
columns_to_keep = [col for col in weather_data.columns if not any(city in col for city in cities_to_remove)]

# Create a new dataframe with the filtered columns
weather_data_filtered = weather_data[columns_to_keep]

# Check the shape to ensure the columns were removed
print(f"Shape of weather_data_filtered: {weather_data_filtered.shape}")
```

```python
# List of observations to be removed
observations_to_remove = ['wind_speed', 'snow_depth']

# Filter out columns that contain the names of the observations to remove
columns_to_keep = [col for col in weather_data_filtered.columns if not any(obs in col for obs in observations_to_remove)]

# Create a new dataframe with the filtered columns
weather_data_cleaned = weather_data_filtered[columns_to_keep]

# Check the shape to ensure the columns were removed
print(f"Shape of weather_data_cleaned: {weather_data_cleaned.shape}")
```

```python
# Add new columns to weather_data_cleaned and copy values from nearby stations

# Copy cloud_cover, pressure, and humidity from Ljubljana to Kassel
weather_data_cleaned['KASSEL_cloud_cover'] = weather_data_cleaned['LJUBLJANA_cloud_cover']
weather_data_cleaned['KASSEL_pressure'] = weather_data_cleaned['LJUBLJANA_pressure']
weather_data_cleaned['KASSEL_humidity'] = weather_data_cleaned['LJUBLJANA_humidity']

# Copy cloud_cover, pressure, and humidity from Munchen to Sonnblick
weather_data_cleaned['SONNBLICK_cloud_cover'] = weather_data_cleaned['MUNCHENB_cloud_cover']
weather_data_cleaned['MUNCHENB_pressure'] = weather_data_cleaned['SONNBLICK_pressure']
weather_data_cleaned['SONNBLICK_humidity'] = weather_data_cleaned['MUNCHENB_humidity']

# Copy cloud_cover, pressure, and humidity from Stockholm to Oslo
weather_data_cleaned['OSLO_cloud_cover'] = weather_data_cleaned['STOCKHOLM_cloud_cover']
weather_data_cleaned['OSLO_pressure'] = weather_data_cleaned['STOCKHOLM_pressure']
weather_data_cleaned['STOCKHOLM_humidity'] = weather_data_cleaned['OSLO_humidity']

# Confirm that the columns have been added successfully
print(weather_data_cleaned[['KASSEL_cloud_cover', 'SONNBLICK_cloud_cover', 'OSLO_cloud_cover']].head())
print(weather_data_cleaned[['KASSEL_pressure', 'SONNBLICK_pressure', 'OSLO_pressure']].head())
print(weather_data_cleaned[['KASSEL_humidity', 'SONNBLICK_humidity', 'OSLO_humidity']].head())
```
