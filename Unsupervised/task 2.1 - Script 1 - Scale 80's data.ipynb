# 2 Scaling

```python
# Select only the numerical columns
numerical_columns = dfyear.select_dtypes(include=['float64']).columns

# Adding the 'DATE' and 'MONTH' columns (ensure these columns exist in the DataFrame)
columns_to_include = ['DATE', 'MONTH'] + list(numerical_columns)

# Subset the dataframe to include numerical data and the 'DATE' and 'MONTH' columns
df_subset = dfyear[columns_to_include]
```

```python
df_subset
```

```python
#create scaler object using StandardScaler from sklearn.preprocessing
#StandardScaler assumes data is normally distributed and scales with a distribution
#around 0 and standard deviation of 1. Scaling happens independently with each variable.

scaler = StandardScaler()
```

```python
# Fit and transform only the numerical columns (excluding 'DATE' and 'MONTH')
df_scaled_values = pd.DataFrame(scaler.fit_transform(df_subset[numerical_columns]), columns=numerical_columns)
```

```python
# Add the 'DATE' and 'MONTH' columns back to the scaled dataframe
df_scaled = pd.concat([df_subset[['DATE', 'MONTH']].reset_index(drop=True), df_scaled_values.reset_index(drop=True)], axis=1)
```
